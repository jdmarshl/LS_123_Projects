### Predicting Medicare Fraud

***
## Introduction

###  Background
Federal statutes allow the U.S. Department of Health and Human Services to exclude individuals and entities from participating in federal health care programs if they have been convicted of fraud or abuse or have engaged in certain other forms of misconduct.
The Health Department maintains a publicly available list of individuals and entities currently excluded from participation in Medicare, Medicaid, and all other federal healthcare programs.
What if we could build a classifier that could detect medical providers that should be added to this list?

### Question
Using publicly available datasets, can a model predict if a medical provider should be excluded from participating in federal healthcare programs?

### Motivation
Medicare fraud is incredibly costly for the government and taxpayers, with losses exceeding hundreds of billions of dollars some years. A model that could detect Medicare fraud would naturally be incredibly useful for the government and/or watchdog groups to catch and prevent these losses. There are tons of publicly available datasets with information about medical providers in the United States that could potentially contain signals to instances of fraud, so I wanted to see whether I could build such a model.

### Feature Selection and Feature Engineering

```python
X = merged.drop(to_drop, axis = 1)
```

```python
X = X.set_index('National Provider Identifier')
X = X[X['Medicare Participation Indicator'] == 'Y'].drop('Medicare Participation Indicator', axis = 1)
X = X[X['Country Code of the Provider']  == 'US'].drop('Country Code of the Provider', axis = 1)
X = X.dropna(subset = suppressed)
X = X.dropna(subset = ['State Code of the Provider'])
```

```python
#fill missing values in the manner described in the EDA section

X['Gender of the Provider'] = X['Gender of the Provider'].fillna(gender_nan_fill)

for col in num_ben:
    X[col] = X[col].fillna(-9)
    X[col] = X[col].apply(lambda x: random.randint(0,11) if x == -9 else x) 
    
for col in perc_ben:
    X[col] = X[col].fillna(-9)
    X[col] = X[col].apply(lambda x: random.randint(75,101) if x == -9 else x)
```

### Resampling 
Perhaps reducing the class imbalance will improve the models' predictions.

```python
pos_class = X[X['excluded'] == 1]
neg_class = X[X['excluded'] == 0]
```

```python
#bootstrap the positive class with a sample size 4 times larger than the population of positive instances

over_sample = pos_class.sample(n = 4*len(pos_class), replace = True)
```

```python
#downsample the negative class to 75% of its original size

under_sample = neg_class.sample(frac = 0.75)
```

```python
#combine resampled sets to end up with a training set that has 25% fewer instances of the negative class 
# and 5 times more instances of the positive class

X_resamp = pos_class.append(under_sample, ignore_index = True).append(over_sample, ignore_index = True)
```

```python
X_resamp.shape
```

```python
#recheck ratio of negative to positive class

targ_count = X_resamp['excluded'].value_counts()
print('Not Excluded:', targ_count[0])
print('Excluded:', targ_count[1])
print('Proportion:', int(round(targ_count[0] / targ_count[1], 0)), ': 1')
```

```python
resamp_bar = X_resamp.groupby('excluded', as_index = False).count().iloc[:, :2]
resamp_bar = resamp_bar.rename({'Gender of the Provider':'count'},  axis =1)
resamp_bar = resamp_bar.replace({0:'excluded', 1:'not excluded'})
ax = resamp_bar.plot.bar(x = 'excluded', y = 'count', legend = False, figsize = (8,10), rot = 0)
plt.yscale('log')
plt.ylabel('count')
plt.xlabel('')
plt.title('Total Instances of Each Target Class After Resampling (log scale)')
i = 0
for p in ax.patches:
    if i == 0:
        i += 1
        ax.annotate(str(p.get_height()), (p.get_x() * 0.4, p.get_height() * 1.07))
    else:
        ax.annotate(str(p.get_height()), (p.get_x() * 1.26, p.get_height() * 1.07))
```

```python
quant_resamp = X_resamp.drop(cat_cols+['excluded'], axis = 1)
cat_resamp = X_resamp.loc[:,cat_cols]
```

```python
#standardize resampled data

scaler = StandardScaler()
resamp_normed = pd.DataFrame(scaler.fit_transform(quant_resamp), columns = quant_resamp.columns, index = quant_resamp.index)
```

```python
X_resamp_processed = resamp_normed.merge(cat_resamp, left_index = True, right_index = True)

#one-hot resampled data
X_resamp_processed = pd.get_dummies(X_resamp_processed)
```

## Conclusion

Accuracy is an effectively useless goodness metric in a problem with such an imbalance in the target classes – a model that predicts only the majority class will achieve a very high accuracy even though we know it isn't doing what we want it to. Usefull goodness metrics in a class imbalanced problem are precision and recall. For this problem, recall is the most important measure since we would rather catch as many cases of fraud as possible even if it means accidentally marking some non-fraudulent cases incorrectly as fraud, though of course we still want to be precise as we can be. In this vein, the decision tree trained on the resampled data far outperformed any of the other models (resampling only made a real difference with the decision tree). 66% recall and 71% precision is honestly quite impressive and it shows that there does indeed appear to be a signal to predict Medicare fraud in the publicly available data I trained on. Of course, this model would need some work before being deployed in the field, but this could certainly serve as a proof of concept. Some ways to begin going about improving the classifier would be using data from more than just one year, collecting more and different data on medical providers (there are loads of potentially useful datasets out there), performing more sophisticated feature selection (like Recursive Feature Elimination) and feature engineering, and employing more advanced resampling techniques such as Tomek links, Cluster Centroids, and SMOTE (Synthetic Minority Oversampling TEchnique). Any search for more data should certainly adhere to legal and ethical standards, and it is important to be cautious in testing and deploying such a tool so as to minimize the likelihood of falsely accusing medical providers of committing fraud.
